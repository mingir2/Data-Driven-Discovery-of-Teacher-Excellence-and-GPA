```
import numpy as np
import pandas as pd
```

```
gpa = pd.read_csv("https://raw.githubusercontent.com/wadefagen/datasets/master/gpa/raw/sp2018.csv", delimiter=',')
gpa.head()
```

```
gpa['Capacity'] = gpa.apply(lambda row: row[['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D', 'D-', 'F', 'W']].sum(), axis = 1)
gpa.head()
```

```
renamed_gpa = gpa.rename(columns={'Course ': 'Course', 'Average Grade': 'GPA', 'Primary Instructor': 'Teacher'})
gpa_tidy = renamed_gpa[['Teacher', 'Course', 'GPA', 'Capacity']]
gpa_tidy.head()
```

```
gpa_tidy[gpa_tidy['Teacher'].isnull()]
```

```
gpa_tidy = gpa_tidy.dropna(subset=['Teacher'])
```

```
tre = pd.read_csv("https://raw.githubusercontent.com/wadefagen/datasets/master/teachers-ranked-as-excellent/csv/tre-sp2018.csv", delimiter=',')
renamed_tre = tre.rename(columns={'lname': 'Teacher', 'ranking': 'Ranking'})
tre_tidy = renamed_tre[['Teacher', 'Ranking']].dropna(subset=['Teacher'])
tre_tidy.head()
```

```
gpa_names = gpa_tidy['Teacher'].unique()
tre_names = tre_tidy['Teacher'].unique()

names = dict()
for tre_name in tre_names.tolist():
  for gpa_name in gpa_names.tolist():
    if tre_name.lower() == gpa_name.split(',')[0].lower():
      names[tre_name] = gpa_name
      
tre_tidy['Teacher'] = tre_tidy['Teacher'].map(names)
tre_tidy = tre_tidy.dropna(subset=['Teacher'])
```

```
data = gpa_tidy.join(tre_tidy.set_index('Teacher'), on ='Teacher')
data['Ranking'].fillna('Good', inplace=True)
data.head(20)
```

```
data['Ranking'].value_counts()
```

```
fig = px.scatter(data, x="GPA", y="Course",
	               size="Capacity", color="Ranking", 
                 marginal_x="histogram",
                 facet_col="Ranking",
                 hover_name="Teacher", size_max=25)
fig.show()
```
